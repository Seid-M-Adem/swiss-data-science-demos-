AI and Automation Pipeline

This notebook demonstrates the AI-driven automation pipeline for processing insurance claims data. It covers loading the synthetic dataset, exploring it, and performing some basic analysis and visualizations.

Step 1: Import Libraries

We need the following libraries to manipulate and visualize the dataset:



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Step 2: Load the Claims Dataset

Next, let's load the synthetic claims data generated by the generate_data.py script.

```python
# Load the claims dataset
data_path = '/workspaces/swiss-data-science-demos-/ai_automation/data/claims_data.csv'
df = pd.read_csv(data_path)

# Display the first 5 rows of the dataset
df.head()
```

Step 3: Overview of the Data

We will now explore the dataset with summary statistics and basic structure information.

```python
# Basic information about the dataset
df.info()
```

```python
# Summary statistics of the dataset
df.describe()
```

Step 4: Data Visualization

Let's create some visualizations to better understand the distribution of the data, including claim amounts, incident types, and processing times.

4.1 Distribution of Claim Amounts

```python
# Plot distribution of claim amounts
plt.figure(figsize=(10,6))
df['claim_amount'].hist(bins=30, color='skyblue')
plt.title('Distribution of Claim Amounts')
plt.xlabel('Claim Amount (CHF)')
plt.ylabel('Frequency')
plt.show()
```

4.2 Number of Claims by Incident Type

```python
# Count of claims by incident type
plt.figure(figsize=(8,6))
df['incident_type'].value_counts().plot(kind='bar', color='orange')
plt.title('Number of Claims by Incident Type')
plt.xlabel('Incident Type')
plt.ylabel('Number of Claims')
plt.show()
```

4.3 Distribution of Claim Processing Time

```python
# Processing time distribution
plt.figure(figsize=(10,6))
df['processing_time'].hist(bins=30, color='green')
plt.title('Distribution of Claim Processing Time')
plt.xlabel('Processing Time (days)')
plt.ylabel('Frequency')
plt.show()
```

Step 5: Filter and Analyze Approved Claims

Let's focus on the approved claims and calculate total claim amounts and average processing time for these.

```python
# Filter for approved claims
approved_claims = df[df['approval_status'] == 'Approved']

# Total claim amount for approved claims
total_approved_claims = approved_claims['claim_amount'].sum()
print(f'Total approved claim amount: CHF {total_approved_claims:.2f}')

# Average processing time for approved claims
avg_processing_time = approved_claims['processing_time'].mean()
print(f'Average processing time for approved claims: {avg_processing_time:.2f} days')
```
